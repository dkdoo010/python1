```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')

import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
```

```python
import numpy as np
def graph(name):
    filt = df['이름'] == name
    df2 = df[filt]
    if len(df2)==0:
        return np.nan
    else:
        df3 = df2.loc[:, ['국어','과학', '사회']]
        df4 = df3.iloc[0]
        plt.bar(df4.index, df4.values)

        return plt 

plt = graph('송태섭')
if pd.isnull(plt):
    print('없음')
else:
    plt.show()
```

```python
name = df['이름']
height = df['키']
```

```python
plt.barh(name, height)
plt.ylabel('학생이름')
plt.xlabel('키')
plt.xlim(160, 210)
plt.show()
```

```python
plt.figure(figsize=(10, 3))
bar = plt.bar(name, height, color='pink', ec='g')
plt.ylim(160, 210)
bar[0].set_hatch('o'), bar[1].set_hatch('/')
for idx, h in enumerate(height):
    plt.text(idx, h+1, h, ha='center')
plt.show()
```

```python
#학생별 평균
# xlabel은 '학생이름',
# ylabel은 '평균점수',
# 학생별 막대그래프위에 평균점수 출력,
# 그래프 제목은 '학생별 평균',
# 그래프 사이즈는 10, 5
```

```python
# 평균 점수 계산
subjects = ['국어', '영어', '수학', '과학', '사회']
df['평균'] = df[subjects].mean(axis=1)

# 그래프 그리기
plt.figure(figsize=(10, 5))
bars = plt.bar(df['이름'], df['평균'], color='pink', edgecolor='black')

# y축 범위 및 눈금 설정
plt.ylim(0, 100)
plt.yticks([y for y in range(0, 105, 5)])

# 라벨 및 제목
plt.xlabel('학생이름')
plt.ylabel('평균점수')
plt.title('학생별 평균')

# 막대 위에 평균 점수 표시
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, f'{height:.1f}', ha='center')

plt.tight_layout()
plt.show()
```
