```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')

import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False)
```

```python
kor = list(df['국어'])
eng = list(df['영어'])
mat = list(df['수학'])
name =list(df['이름'])
```

```python
plt.figure(figsize=(10,3))
plt.plot(name, kor, label='국어', color='brown', marker='o', ls='--')
plt.plot(name, eng, label='영어', color='black', marker='s', ls=':')
plt.plot(name, mat, label='수학', color='purple', marker='X')
plt.legend(loc=(1.02, 0.5))
for idx, k in enumerate(kor):
    plt.text(name[idx], kor[idx]+2, k, color='brown', ha='center', size=8)
for idx, e in enumerate(eng):
    plt.text(name[idx], eng[idx]+2, e, color='black', ha='center', size=8)
for idx, m in enumerate(mat):
    plt.text(name[idx], mat[idx]+2, m, color='purple', ha='center', size=8)
plt.show()
```
