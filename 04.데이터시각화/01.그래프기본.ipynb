```python
import matplotlib.pyplot as plt
```

```python
x = [-1, 2, 3]
y = [2, 4, 8]
plt.plot(x, y)
plt.show
```

```python
#한글설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 20)
plt.rc('axes', unicode_minus = False)
```

```python
import matplotlib.font_manager as fm
flist = fm.fontManager.ttflist
[f.name for f in flist]
```

```python
plt.plot(x, y)
plt.title('꺽은선그래프')
plt.show
```

```python
plt.title('꺽은선그래프', size = 20)
plt.plot(x,y)
plt.rc('font', size=8)
plt.xlabel('X축', color='purple', size = 15, loc = 'right')
plt.ylabel('Y축', color='brown', size = 15, loc = 'top')
plt.show()
```

```python
plt.plot(x, y)
plt.xticks(x)
plt.yticks([y for y in range(0, 10, 2)])
plt.show()
```
