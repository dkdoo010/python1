```python
with open('영어가사.text', 'r') as file:
    text = file.read()
print(text)
```

```python
from wordcloud import WordCloud
import matplotlib.pylab as plt
```

```python
wc = WordCloud().generate(text)
plt.imshow(wc)
plt.axis('off')
plt.savefig('영어가사.png')
plt.show()
```

```python
import numpy as np
from PIL import Image
```

```python
heart_mask = np.array(Image.open('heart_mask.png'))
```

```python
wc = WordCloud(
    width=800,
    height=800,
    background_color='pink',
    min_font_size=20,
    mask=heart_mask
)
wc.generate(text)
```

```python
plt.figure(figsize=(5,5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.savefig('골든하트.png')
plt.show()
```

```python
import jpype
from konlpy.tag import Okt
okt = Okt()

```

```python
with open('애국가.txt', 'r', encoding='utf-8') as file:
    text = file.read()
    print(text)
```

```python
okt = Okt()
nouns = okt.nouns(text)
nouns = [noun for noun in nouns if len(noun) > 1]
```

```python
counter = Counter(nouns)
counter
```

```python
import numpy as np
from PIL import Image
heart_mask = np.array(Image.open('heart_mask.png'))
```

```python
wc = WordCloud(
    width=800,
    height=800,
    font_path='C:/Windows/Fonts/H2HDRM.TTF',
    scale=2.0,
    max_font_size=250,
    mask=heart_mask,
    background_color='white'
).generate_from_frequencies(counter)
```

```python
plt.imshow(wc)
plt.axis('off')
plt.savefig('애국가.png')
plt.show()
```
