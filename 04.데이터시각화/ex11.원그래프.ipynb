```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun gothic')
plt.rc('axes', unicode_minus=False)

import pandas as pd
df = pd.read_csv('data/score.csv')
df
```

```python
group = df.groupby('SW특기').size()
group
```

```python
labels = group.index
values = group.values
labels, values
```

```python
plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=True)
plt.show()
```

```python
values=[30, 25, 20, 13, 10, 2]
labels =['Python', 'Java', 'Javascript', 'c#', 'c/c++', 'ETC' ]
```

```python
explode = [0.1] * len(values)  # 적당한 강조 폭 설정

plt.pie(values, labels=labels, explode=explode, autopct='%.1f%%',
        startangle=90, counterclock=False)

plt.legend(loc=(1.2, 0.5))
plt.title('SW특기별 학생 분포')
plt.show()

```

```python
def custom_autopct(pct):
    if pct <=2:
        return ''
    else:
        return f'{pct:.0f}%'
```

```python
explode = [0.1] * len(values)  # 살짝만 띄우기
colors = ['#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF']

def custom_autopct(pct):
    return f'{pct:.1f}%'

plt.figure(figsize=(7, 7))
plt.pie(values, labels=labels, explode=explode, colors=colors,
        wedgeprops={'width':0.8, 'edgecolor':'white', 'linewidth':5},
        autopct=custom_autopct, startangle=90, counterclock=False)

plt.legend(loc=(1.2, 0.5))
plt.show()

```
