```python
import pandas as pd
score= pd.read_csv('data/score.csv')
score.rename(columns={'Unnamed: 0': '지원번호'}, inplace=True)
score.set_index('지원번호', inplace=True)
df_score = score.loc[:, '국어':'사회']
df_score
```

```python
df = df_score.copy()
df.reset_index(inplace=True)
df['지원번호']=df['지원번호'].str.replace('번', '')
df.set_index('지원번호', inplace=True)
df.to_csv('data/학생성적.csv')
```
